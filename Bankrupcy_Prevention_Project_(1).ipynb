```python
#inpute variable
x=df.iloc[:,:6]
#target variable
y=df['class']
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
ss_x=ss.fit_transform(x)
ss_x=pd.DataFrame(ss_x)
ss_x
```

```python
#Splitting data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(ss_x,y,random_state=0,test_size=25)
```

```python
#preparing models
#=================================================1.Logistic Regression===========================================================
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(x_train,y_train)
```

```python
y_pred_train=LR.predict(x_train)
y_pred_test=LR.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
#train Accuracy
ac_train_LR=accuracy_score(y_train,y_pred_train)
ac_train_LR
```

```python
#test Accuracy
ac_test_LR=accuracy_score(y_test,y_pred_test)
ac_test_LR
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score
#Training dataset
cm1=confusion_matrix(y_train,y_pred_train)
cm1
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score
cm2=confusion_matrix(y_test,y_pred_test)
cm2
```

```python
rs1_LR=recall_score(y_train,y_pred_train)
rs1_LR
```

```python
rs2_LR=recall_score(y_test,y_pred_test)
rs2_LR
```

```python
TN=cm1[0,0]
FP=cm1[0,1]

TNR1=TN/(TN+FP)
print("Specificity:",TNR1.round(2))
```

```python
TN=cm2[0,0]
FP=cm2[0,1]

TNR2=TN/(TN+FP)
print("Specificity:",TNR2.round(2))
```

```python
from sklearn.metrics import precision_score,f1_score
print("Precision_Score_LR(Training):",precision_score(y_train,y_pred_train).round(2))
precision_LR=precision_score(y_test,y_pred_test)
print("Precision_Score_LR(Testing):",precision_LR.round(2))
```

```python
from sklearn.metrics import precision_score,f1_score
print("F1_Score_LR(Training):",f1_score(y_train,y_pred_train).round(2))
F1_LR=f1_score(y_test,y_pred_test)
print("F1_Score_LR(Testing):",F1_LR.round(2))
```

```python
df['y_predicted_proba']=LR.predict_proba(x)[:,1]
df.head()
```
